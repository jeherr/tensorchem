```python
import scripts.mset_scope as MS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
meta_msets, meta_mset_natoms, meta_elements = MS.get_meta_mset_info()
```

```python
ani_msets, ani_mset_natoms, ani_elements = MS.get_ani_mset_info()
```

```python
print("Total number of unique chemspider metadynamics molecules: ", len(meta_msets))
print("Total number of unique molecules in ANI-1x dataset: ", len(ani_msets))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(13.5, 4.5)

ax1.hist(meta_mset_natoms, bins=100, edgecolor='k', range=[0,100])
ax1.set_xlabel("Number of Atoms per Unique Molecule")
ax1.set_ylabel("Number of Unique Meta Molecules")
ax1.set_title('Histogram of Atoms per Unique Meta Molecule')

ax2.hist(ani_mset_natoms, bins=70, edgecolor='k', range=[0,70])
ax2.set_xlabel("Number of Atoms per Unique  Molecule")
ax2.set_ylabel("Number of Unique ANI1x Molecules")
ax2.set_title('Histogram of Atoms per Unique ANI1x Molecule')

fig.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(10.5, 4.5)

ax1.bar(list(meta_elements.keys()), meta_elements.values())
ax1.set_title('Bar Chart of Elements in Meta MSets')
ax1.set_yscale("log")

ax2.bar(list(ani_elements.keys()), ani_elements.values())
ax2.set_title('Bar Chart of Elements in ANI-1x MSets')
ax2.set_yscale("log")

fig.tight_layout()
plt.show()
```
